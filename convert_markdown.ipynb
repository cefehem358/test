# Convert Markdown Files to PDF

This notebook searches the current working directory for all `.md` files and converts each to a PDF. It uses `pandoc` via `subprocess` (or `pypandoc` if available). Optionally, generated PDFs can be merged into a single document.

```python
# Import Required Libraries
import os
import glob
import subprocess

try:
    import pypandoc
    has_pypandoc = True
except ImportError:
    has_pypandoc = False

print("pypandoc available:", has_pypandoc)
```

## Locate Markdown Files

Find all files with the `.md` extension in the current directory (non-recursive by default).

```python
md_files = glob.glob('*.md')
print("Markdown files found:", md_files)
```

## Convert Markdown to PDF

Iterate through the list and convert each Markdown file to a PDF using either `pypandoc` or calling `pandoc` directly.

```python
for md in md_files:
    pdf = os.path.splitext(md)[0] + '.pdf'
    print(f"Converting {md} -> {pdf}")
    if has_pypandoc:
        pypandoc.convert_file(md, 'pdf', outputfile=pdf)
    else:
        # call pandoc; this requires pandoc installed in environment
        subprocess.run(['pandoc', md, '-o', pdf], check=False)

print("Conversion complete.")
```

## Combine PDFs (Optional)

If you'd like to merge all generated PDFs into a single file, you could use a library like PyPDF2.

```python
try:
    from PyPDF2 import PdfMerger
    merger = PdfMerger()
    for md in md_files:
        pdf = os.path.splitext(md)[0] + '.pdf'
        if os.path.exists(pdf):
            merger.append(pdf)
    merger.write('combined.pdf')
    merger.close()
    print('Combined PDF created as combined.pdf')
except ImportError:
    print('PyPDF2 not installed; skipping combine step')
```
